man.py

```
class Man:

  def __init__(self, name):
    self.name = name
    print("Initialized!")

  def hello(self):
    print("Hello " + self.name + "!")

  def goodbye(self):
    print("Good-bye " + self.name + "!")

m = Man("David") # クラスの初期化
m.hello()
m.goodbye()
```

simple_graph.py

```
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(0, 6, 0.1) # 0~6、[0, 0.1, 0.2, ..., 6]
y = np.sin(x)

plt.plot(x, y)
plt.show()
```

sin_cos_graph.py

```
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(0, 6, 0.1)
y1 = np.sin(x)
y2 = np.cos(x)

plt.plot(x, y1, label="sin")
plt.plot(x, y2, linestyle="--", label="cos")
plt.xlabel("x")
plt.ylabel("y")
plt.title("sin & cos")
plt.legend()
plt.show()
```
